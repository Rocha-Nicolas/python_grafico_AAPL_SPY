```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

#Descargar los datos de Yahoo Finance para AAPL y SPY500 (.yf los devuelve en formato dataFrame)
appl = yf.download('AAPL', start='2018-01-01', end='2023-04-06')
spy = yf.download('SPY', start='2018-01-01', end='2023-04-06')

#Configuracion del tamao del gracifo, titulo
fig, ax1= plt.subplots(figsize = (11,5))
plt.suptitle("SPY y AAPL | 2018 - 2023 | Cada COVID-19 y subida de tipos de inters", y = 1.05)
plt.style.use('fivethirtyeight')


fecha_caida_inicio = '2020-02-08'
fecha_caida_fin = '2020-03-24'

#Carga en el grafico con los precios del S&P 500 y de APPL
ax1.plot(appl.index, appl['Adj Close'],lw=1, ls="--", color='grey', label='AAPL')
ax1.plot(spy.index, spy['Adj Close'], lw=1,ls="-",color='k', label='SPY')

#Pintado de el mometno de la caida de la bolsa por el COVID-19
cituacion = ((spy.index >= fecha_caida_inicio) & (spy.index <= fecha_caida_fin))
ax1.fill_between(spy.index, spy["Adj Close"].max(), where=cituacion, color="violet", alpha=0.3)

#Marca con una linea vertical en el grafico marcando el inicio de la subida de tipos de interes
fecha_subida = pd.to_datetime('2022-03-16')
ax1.axvline(x = fecha_subida, ymin=0, c='red', ls='--', lw=1)


#Leyenda
plt.legend(["APPL","S&P 500","Caida COVID-19","Subida de Tipos"], loc="upper left")

plt.show()
```
